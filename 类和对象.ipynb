```python
class human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def sleep(self):
        print("睡着了")
   
human1=human(input(), int(input()))
human2=human("小明", 20)
human1.sleep()
print(human1.name,human1.age)
print(human2.name,human2.age)
```
